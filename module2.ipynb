```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
start = datetime.datetime(2010,1,1)
end = datetime.datetime(2021,1,1)
```

```python
tesla = web.DataReader("TSLA", 'yahoo', start, end)
ford = web.DataReader("F", 'yahoo', start, end)
gm = web.DataReader("GM", 'yahoo', start, end)
```

```python
tesla.to_csv('Tesla_Stock.csv')
ford.to_csv('Ford_Stock.csv')
gm.to_csv('GM_Stock.csv')
```

```python
tesla['Open'].plot(label = 'Tesla', figsize = (15,7))
ford['Open'].plot(label = 'Ford')
gm['Open'].plot(label = 'General Motors')
plt.title("Stock Prices of Tesla(Blue), Ford(Orange) and General Motors(Green)")
plt.ylabel('Stock Price')
plt.legend()
```

```python
tesla['Volume'].plot(label = 'Tesla', figsize = (15,7))
ford['Volume'].plot(label = 'Ford')
gm['Volume'].plot(label = 'GM')
plt.title('Volume Traded of Tesla, Ford and General Motors')
plt.ylabel('Volume Traded')
plt.legend()
```

```python
ford.iloc[[ford['Volume'].argmax()]]
```

```python
ford.iloc[460:53]['Open'].plot()label = 'Ford Share Price', figsize = (15,7))
plt.title('Ford Stock Analysis (last quarter of 2011')
plt.ylabel('Stock Price')
plt.legend()
# There's fall in stock price of Ford during end of 4th Quarter of 2011
```

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
gm['Total Traded'] = gm['Open'] * gm['Volume']
# Formula of total amount of money traded is open pricing times volume
```
